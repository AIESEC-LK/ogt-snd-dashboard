```python
import pandas as pd
```

```python
app = pd.read_csv('jan23_applications.csv')
app
```

```python
#remobe GV (New)
# remove rows by filtering
app = app[app['Product'] != 'GV (New)']
# display the dataframe
app
```

```python
app_final = app.drop(columns=['SDG Goal',
                  'SDG Target',
                  'utm_term',
                  'utm_medium',
                  'utm_source',
                  'utm_content',
                  'utm_campaign',
                  'Tags',
                  'Global Project',
                  'Global Partner'])

app_final
```

```python
pbg = pd.read_csv('jan23_People-backgrounds.csv')
pbg
```

```python
ps = pd.read_csv('jan23_People-skills.csv')
ps
```

```python
app_pbg = pd.merge(app_final, pbg, on = ["EP ID","Home LC","Home MC","Home Region"], how = "left")

app_pbg
```

```python
app_pbg_ps = pd.merge(app_pbg, ps, on = ["EP ID","Home LC","Home MC","Home Region"], how = "left")

app_pbg_ps
```

```python
app_pbg_ps = app_pbg_ps[app_pbg_ps['Home MC'] == 'Sri Lanka']
# display the dataframe
app_pbg_ps
```

```python
app_pbg_ps.to_csv('application_people_skills_background.csv')
```
